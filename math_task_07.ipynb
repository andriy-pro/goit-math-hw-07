## Завдання 1

> У партії з 50 деталей 5 нестандартних. Визнач ймовірність того, що серед обраних навмання для перевірки шести деталей дві виявляться нестандартними. Розв’яжи аналітично.
> *Рахувати конкретне число не потрібно*

### Розв'язання

Задачу можна розв'язати за допомогою комбінаторики та формули ймовірності класичної схеми.
Для визначення ймовірності, що серед обраних навмання шести деталей дві виявляться нестандартними, ми можемо використовувати формулу для обчислення ймовірності в комбінаційних задачах.

## Крок 1: Визначення параметрів задачі

- Загальна кількість деталей (N): 50
- Кількість нестандартних деталей (K): 5
- Кількість стандартних деталей: $ N - K = 50 - 5 = 45 $
- Кількість обраних деталей (n): 6
- Кількість нестандартних деталей серед обраних (k): 2

## Крок 2: Використання формули ймовірності

Ймовірність того, що серед обраних деталей буде k нестандартних, можна обчислити за допомогою формули комбінацій:

$$
P(X = k) = \frac{\binom{K}{k} \cdot \binom{N-K}{n-k}}{\binom{N}{n}},
$$

де:

- $\binom{K}{k}$ — кількість способів вибрати k нестандартних деталей з K,

- $\binom{N-K}{n-k}$ — кількість способів вибрати $ n-k $ стандартних деталей з $ N-K $,

- $\binom{N}{n}$ — загальна кількість способів вибрати n деталей з N.

## Крок 3: Підставлення значень у формулу

Тепер підставимо значення у формулу:

1. $\binom{K}{k} = \binom{5}{2}$ — кількість способів вибрати 2 нестандартні деталі з 5.

2. $\binom{N-K}{n-k} = \binom{45}{4}$ — кількість способів вибрати 4 стандартні деталі з 45.

3. $\binom{N}{n} = \binom{50}{6}$ — загальна кількість способів вибрати 6 деталей з 50.

## Крок 4: Обчислення комбінацій

Тепер можемо записати ймовірність:

$$
P(X = 2) = \frac{\binom{5}{2} \cdot \binom{45}{4}}{\binom{50}{6}}
$$

## Крок 5: Визначення значень комбінацій

1. $\binom{5}{2} = \frac{5!}{2!(5-2)!} = \frac{5 \cdot 4}{2 \cdot 1} = 10$

2. $\binom{45}{4} = \frac{45!}{4!(45-4)!} = \frac{45 \cdot 44 \cdot 43 \cdot 42}{4 \cdot 3 \cdot 2 \cdot 1} = 148995$

3. $\binom{50}{6} = \frac{50!}{6!(50-6)!} = \frac{50 \cdot 49 \cdot 48 \cdot 47 \cdot 46 \cdot 45}{6 \cdot 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1} = 15890700$

## Крок 6: Підставлення значень у формулу

Тепер можемо підставити значення у формулу:

$$
P(X = 2) = \frac{10 \cdot 148995}{15890700}
$$

## Висновок

Таким чином, ймовірність того, що серед обраних навмання шести деталей дві виявляться нестандартними, дорівнює:

$$
P(X = 2) = \frac{10 \cdot 148995}{15890700}
$$

Це вираження можна спростити до конкретного числового значення, якщо це необхідно, але в умові задачі зазначено, що рахувати конкретне число не потрібно. Тому ми залишаємо відповідь у вигляді формули.
