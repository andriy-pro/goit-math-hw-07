## Завдання 1

> У партії з 50 деталей 5 нестандартних. Визнач ймовірність того, що серед обраних навмання для перевірки шести деталей дві виявляться нестандартними. Розв’яжи аналітично.
> *Рахувати конкретне число не потрібно*

### Розв'язання

Задачу можна розв'язати за допомогою комбінаторики та формули ймовірності класичної схеми.
Для визначення ймовірності, що серед обраних навмання шести деталей дві виявляться нестандартними, ми можемо використовувати формулу для обчислення ймовірності в комбінаційних задачах.

#### Крок 1: Визначення параметрів задачі

- Загальна кількість деталей (N): 50
- Кількість нестандартних деталей (K): 5
- Кількість стандартних деталей: $ N - K = 50 - 5 = 45 $
- Кількість обраних деталей (n): 6
- Кількість нестандартних деталей серед обраних (k): 2

#### Крок 2: Використання формули ймовірності

Ймовірність того, що серед обраних деталей буде k нестандартних, можна обчислити за допомогою формули комбінацій:

$$
P(X = k) = \frac{\binom{K}{k} \cdot \binom{N-K}{n-k}}{\binom{N}{n}},
$$

де:

- $\binom{K}{k}$ — кількість способів вибрати k нестандартних деталей з K,

- $\binom{N-K}{n-k}$ — кількість способів вибрати $ n-k $ стандартних деталей з $ N-K $,

- $\binom{N}{n}$ — загальна кількість способів вибрати n деталей з N.

#### Крок 3: Підставлення значень у формулу

Тепер підставимо значення у формулу:

1. $\binom{K}{k} = \binom{5}{2}$ — кількість способів вибрати 2 нестандартні деталі з 5.

2. $\binom{N-K}{n-k} = \binom{45}{4}$ — кількість способів вибрати 4 стандартні деталі з 45.

3. $\binom{N}{n} = \binom{50}{6}$ — загальна кількість способів вибрати 6 деталей з 50.

#### Крок 4: Обчислення комбінацій

Тепер можемо записати ймовірність:

$$
P(X = 2) = \frac{\binom{5}{2} \cdot \binom{45}{4}}{\binom{50}{6}}
$$

#### Крок 5: Визначення значень комбінацій

1. $\binom{5}{2} = \frac{5!}{2!(5-2)!} = \frac{5 \cdot 4}{2 \cdot 1} = 10$

2. $\binom{45}{4} = \frac{45!}{4!(45-4)!} = \frac{45 \cdot 44 \cdot 43 \cdot 42}{4 \cdot 3 \cdot 2 \cdot 1} = 148995$

3. $\binom{50}{6} = \frac{50!}{6!(50-6)!} = \frac{50 \cdot 49 \cdot 48 \cdot 47 \cdot 46 \cdot 45}{6 \cdot 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1} = 15890700$

#### Крок 6: Підставлення значень у формулу

Тепер можемо підставити значення у формулу:

$$
P(X = 2) = \frac{10 \cdot 148995}{15890700}
$$

### **Відповідь**

Таким чином, ймовірність того, що серед обраних навмання шести деталей дві виявляться нестандартними, дорівнює:

$$
P(X = 2) = \frac{10 \cdot 148995}{15890700}
$$

Це вираження можна спростити до конкретного числового значення, якщо це необхідно, але в умові задачі зазначено, що рахувати конкретне число не потрібно. Тому ми залишаємо відповідь у вигляді формули.

## Завдання 2

> В тирі стрілець проводить 7 пострілів по мішені з ймовірністю влучання кожного 0.8. Яка ймовірність того, що буде не менше за 5 влучань?

### Розв'язання

Задачу можна розв'язати за допомогою [біноміального розподілу](https://uk.wikipedia.org/wiki/Біноміальний_розподіл). У даному випадку ми маємо 7 пострілів, і ймовірність влучання в мішень для кожного пострілу дорівнює 0.8. Нам потрібно знайти ймовірність того, що буде не менше ніж 5 влучань.

#### Крок 1: Визначення параметрів задачі

- **Кількість пострілів (n):** 7
- **Ймовірність влучання (p):** 0.8
- **Кількість влучань (k):** 5, 6, 7 (оскільки ми шукаємо ймовірність не менше ніж 5)

#### Крок 2: Використання формули біноміального розподілу

Ймовірність того, що з $ n $ незалежних випробувань буде $ k $ успішних результатів, можна обчислити за допомогою формули біноміального розподілу:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k},
$$

де:

- $ \binom{n}{k} $ — кількість способів вибрати $ k $ успішних результатів з $ n $,

- $ p^k $ — ймовірність успіху,

- $ (1-p)^{n-k} $ — ймовірність невдачі.

#### Крок 3: Обчислення ймовірностей для k = 5, 6, 7

Тепер обчислимо ймовірності для кожного з випадків:

1. Для $ k = 5 $:
   $$
   P(X = 5) = \binom{7}{5} (0.8)^5 (0.2)^2
   $$

2. Для $ k = 6 $:
   $$
   P(X = 6) = \binom{7}{6} (0.8)^6 (0.2)^1
   $$

3. Для $ k = 7 $:
   $$
   P(X = 7) = \binom{7}{7} (0.8)^7 (0.2)^0
   $$

#### Крок 4: Обчислення комбінацій

1. $ \binom{7}{5} = \frac{7!}{5!(7-5)!} = \frac{7 \times 6}{2 \times 1} = 21 $

2. $ \binom{7}{6} = \frac{7!}{6!(7-6)!} = 7 $

3. $ \binom{7}{7} = 1 $

#### Крок 5: Підставлення значень у формули

Тепер підставимо значення у формули:

1. Для $ k = 5 $:
   $$
   P(X = 5) = 21 \cdot (0.8)^5 \cdot (0.2)^2 = 21 \cdot 0.32768 \cdot 0.04 = 0.27525
   $$

2. Для $ k = 6 $:
   $$
   P(X = 6) = 7 \cdot (0.8)^6 \cdot (0.2)^1 = 7 \cdot 0.262144 \cdot 0.2 = 0.367
   $$

3. Для $ k = 7 $:
   $$
   P(X = 7) = 1 \cdot (0.8)^7 \cdot (0.2)^0 = 0.2097152
   $$

#### Крок 6: Обчислення загальної ймовірності

Тепер обчислимо загальну ймовірність того, що буде не менше ніж 5 влучань:

$$
P(X \geq 5) = P(X = 5) + P(X = 6) + P(X = 7)
$$

$$
P(X \geq 5) = 0.27525 + 0.367 + 0.2097152 = 0.8519652
$$

### **Відповідь**

Таким чином, ймовірність того, що буде не менше ніж 5 влучань, дорівнює:

$$
P(X \geq 5) \approx 0.852
$$

Це означає, що стрілець має приблизно **85.2%** ймовірності влучити не менше ніж 5 разів з 7 пострілів.

## Завдання 3

> Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12? Розв’яжи програмно, згенерувавши всі можливі комбінації.

```python
from itertools import product # Ця функція дозволяє генерувати декартовий добуток (усі можливі комбінації) для заданих ітерацій.

def count_combinations_and_probabilities(target_sums):
    # Створюємо словник для зберігання кількості комбінацій для кожної цільової суми.
    # Словник має ключі, які відповідають цільовим сумам, і значення, які спочатку дорівнюють 0.
    counts = {target: 0 for target in target_sums}
    
    # Генеруємо всі можливі комбінації для трьох кубиків.
    # Для кожного кубика можуть бути значення від 1 до 6.
    # Використовуємо product для створення всіх можливих комбінацій.
    for combo in product(range(1, 7), repeat=3):
        # Обчислюємо суму значень у поточній комбінації.
        combo_sum = sum(combo)
        # Перевіряємо, чи дорівнює сума одній з цільових сум.
        if combo_sum in counts:
            # Якщо так, збільшуємо лічильник для цієї суми на 1.
            counts[combo_sum] += 1
            
    # Обчислюємо загальну кількість комбінацій.
    # Оскільки у нас 3 кубики, кожен з яких має 6 граней, загальна кількість комбінацій дорівнює 6^3 = 216.
    total_combinations = 6 ** 3  # 6^3 = 216

    # Обчислюємо ймовірності для кожної цільової суми.
    # Ймовірність = (кількість комбінацій для суми) / (загальна кількість комбінацій).
    # Використовуємо словник для зберігання ймовірностей.
    probabilities = {target: count / total_combinations for target, count in counts.items()}
    
    # Повертаємо два словники: counts (кількість комбінацій) та probabilities (ймовірності).
    return counts, probabilities

# Визначаємо цільові суми, для яких будемо підраховувати комбінації та ймовірності.
target_sums = [11, 12]

# Викликаємо функцію count_combinations_and_probabilities з нашими цільовими сумами.
# Зберігаємо результати в змінні combinations (кількість комбінацій) та probabilities (ймовірності).
combinations, probabilities = count_combinations_and_probabilities(target_sums)

for target in target_sums:
    print(f"Сума: {target}") 
    print(f"Кількість комбінацій: {combinations[target]}")
    print(f"Ймовірність: {probabilities[target]:.4f}")
    print()
```

## Завдання 4

> Схема електричного ланцюга наведена на рисунку. Через ділянку схеми, що вийшла з ладу, струм не проходить. Нехай подія $ A_i $ — вихід з ладу елемента $i$, $i = \overline{1, 6}$. Виразіть подію $ A $ й $\overline{A}$ через події $ A_i $, якщо $ A $ — вихід з ладу всієї схеми. Вирішіть аналітично.
> 
> ![Схема](https://s3.eu-north-1.amazonaws.com/lms.goit.files/1ac1d94a-dcd1-4a3a-aa22-d0da1c34b2f320.png)

### Розв'язання

#### Опис схеми

1. **Перший та другий елементи ($A_1$, $A_2$):** ці два елементи з'єднані паралельно. Це означає, що для того, щоб їх частина схеми вийшла з ладу, обидва елементи повинні вийти з ладу. 
   - **Подія**: Вихід з ладу обох елементів $A_1$ та $A_2$ можна описати як $A_1 \land A_2$.
   
2. **Третій та четвертий елементи ($A_3$, $A_4$):** ці елементи з'єднані послідовно. Вихід з ладу хоча б одного з них призведе до виходу з ладу цієї **підсхеми** *(не всієї схеми)*. 
   - **Подія**: Вихід з ладу підсхеми можна описати як $A_3 \lor A_4$.
   
3. **П'ятий елемент ($A_5$):** цей елемент з'єднаний паралельно з підсхемою $A_3$ та $A_4$. Для того, щоб вийшла з ладу вся частина схеми, потрібно, щоб вийшли з ладу елементи $A_5$ та підсхема $A_3 \lor A_4$.
   - **Подія**: Вихід з ладу цієї частини схеми можна описати як $A_5 \land (A_3 \lor A_4)$.
   
4. **Шостий елемент ($A_6$):** цей елемент з'єднаний послідовно з усією схемою. Якщо цей елемент вийде з ладу, уся схема вийде з ладу незалежно від стану інших елементів.
   - **Подія**: Вихід з ладу цього елемента можна описати як $A_6$.

#### Подія виходу з ладу всієї схеми ($A$)

Вихід з ладу всієї схеми відбудеться в двох випадках:
- Виходять з ладу елементи $A_1$ та $A_2$, шостий елемент $A_6$, і вся частина схеми з $A_5$ та $A_3 \lor A_4$.
- Тобто, вихід з ладу всіх частин схеми.

Отже, вихід з ладу всієї схеми можна описати як:

$$
A = (A_1 \land A_2) \lor (A_5 \land (A_3 \lor A_4)) \lor A_6
$$

#### Подія "невиходу з ладу всієї схеми" ($ \neg A $)

Невихід з ладу всієї схеми можна описати через заперечення кожного з компонентів, тобто:

$$
\neg A = \neg[(A_1 \land A_2) \lor (A_5 \land (A_3 \lor A_4)) \lor A_6] = \neg(A_1 \land A_2) \land \neg(A_5 \land (A_3 \lor A_4)) \land \neg A_6
$$

Також можна використати *[правила де Моргана](https://uk.wikipedia.org/wiki/Правила_де_Моргана)* для спрощення цього виразу:

$$
\neg A = (\neg A_1 \lor \neg A_2) \land (\neg A_5 \lor \neg(A_3 \lor A_4)) \land \neg A_6
$$

### **Відповідь**

Вихід з ладу всієї схеми:
$ A = (A_1 \land A_2) \lor (A_5 \land (A_3 \lor A_4)) \lor A_6 $

Невихід з ладу:
$ \neg A = (\neg A_1 \lor \neg A_2) \land (\neg A_5 \lor \neg(A_3 \lor A_4)) \land \neg A_6 $

- Площа кільця: $ S_{\text{кільця}} = S_1 - S_2 = 100\pi - 25\pi = 75\pi \, \text{см}^2 $

#### 3. Розрахунок ймовірності:

$$
P(A) = \frac{S_{\text{кільця}}}{S_1} = \frac{75\pi \, \text{см}^2}{100\pi \, \text{см}^2} = \frac{75}{100} = \frac{3}{4} = 0.75

### **Відповідь**

Ймовірність того, що точка, кинута навмання у велике коло, потрапить у кільце, утворене побудованими окружностями, дорівнює $\frac{3}{4}$, або 0.75 (75%).

## Завдання 6

> Урна містить одну кулю, про яку відомо, що вона або біла, або чорна з однаковими ймовірностями. В урну кладуть білу кулю і потім навмання виймають одну кулю. Вона виявилася білою. Яка ймовірність того, що куля, яка залишилася, є білою? Розв’яжи аналітично.

### Розв'язання

Ця задача стосується умовної ймовірності, і її можна вирішити за допомогою [**теореми Баєса**](https://uk.wikipedia.org/wiki/Теорема_Баєса).

**Теорема Баєса** дозволяє обчислити умовну ймовірність події на основі попередніх знань про умови, пов'язані з цією подією. Формула теореми Баєса виглядає так:

$$
P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)}
$$

де:
- $ P(B|A) $ — ймовірність події $ B $ за умови, що подія $ A $ вже відбулася.
- $ P(A|B) $ — ймовірність події $ A $ за умови, що подія $ B $ вже відбулася.
- $ P(B) $ — ймовірність події $ B $.
- $ P(A) $ — ймовірність події $ A $.


#### 1. Визначення подій

- Нехай $A$ — подія, що витягнута біла куля.
- Нехай $B_1$ — подія, що спочатку в урні була біла куля.
- Нехай $B_2$ — подія, що спочатку в урні була чорна куля.

Нам потрібно знайти ймовірність того, що залишилася біла куля, за умови, що витягнули білу кулю, тобто $P(B_1 | A)$.

#### 2. Початкові ймовірності

- Спочатку ймовірність того, що в урні була біла куля: $P(B_1) = 0.5$.
- Ймовірність того, що в урні була чорна куля: $P(B_2) = 0.5$.

#### 3. Умовні ймовірності

- Якщо в урні була біла куля ($B_1$), то після додавання ще однієї білої кулі ймовірність витягти білу кулю дорівнює:
$$
P(A | B_1) = \frac{2}{2} = 1
$$

- Якщо в урні була чорна куля ($B_2$), то після додавання білої кулі ймовірність витягти білу кулю дорівнює:
$$
P(A | B_2) = \frac{1}{2} = 0.5
$$

#### 4. Використання теореми Баєса

Застосуємо формулу Баєса для обчислення ймовірності $P(B_1 | A)$:
$$
P(B_1 | A) = \frac{P(A | B_1) \cdot P(B_1)}{P(A)}
$$

Де $P(A)$ обчислюється як загальна ймовірність витягти білу кулю:
$$
P(A) = P(A | B_1) \cdot P(B_1) + P(A | B_2) \cdot P(B_2)
$$

Підставляємо значення:
$$
P(A) = 1 \cdot 0.5 + 0.5 \cdot 0.5 = 0.5 + 0.25 = 0.75
$$

Тепер знайдемо $P(B_1 | A)$:
$$
P(B_1 | A) = \frac{1 \cdot 0.5}{0.75} = \frac{0.5}{0.75} = \frac{2}{3}
$$

### **Відповідь**

Таким чином, ймовірність того, що залишилася біла куля, за умови, що витягнули білу кулю, дорівнює $\frac{2}{3}$, або приблизно 0.67 (67%).

## Завдання 7: "Випадкове блукання"

> Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. У початковий момент часу $ t = 0 $ ціна рівна 0. Визнач середню ціну акції через 3 одиниці часу ($ t = 3 $).
> 
> а) Розв’яжи аналітично.

### Розв'язання (а)

#### Визначення параметрів:
- Ціна може збільшитись на 2 або зменшитись на 1 з рівною ймовірністю.
- Початкова ціна = 0.
- Нас цікавить середня ціна через 3 одиниці часу.

#### Розрахунок:
- За один крок очікувана зміна ціни:
  E(зміна) = 0.5 * 2 + 0.5 * (-1) = 1 - 0.5 = 0.5

- За 3 кроки очікувана зміна:
  E(зміна за 3 кроки) = 3 * 0.5 = 1.5

- Оскільки початкова ціна 0, середня ціна через 3 одиниці часу:
  E(ціна при t=3) = 0 + 1.5 = 1.5

### **Відповідь (a)**

Середня ціна акції через 3 одиниці часу дорівнює 1.5.

---

### Розв'язання (б)

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_price(steps=3):
    changes = np.random.choice([2, -1], size=steps)
    return changes.sum()

def run_simulations(n_sims):
    return [simulate_price() for _ in range(n_sims)]

# Симуляції
sims = [10, 100, 1000, 10000]
results = {n: run_simulations(n) for n in sims}

# Розрахунок середніх
means = {n: np.mean(res) for n, res in results.items()}

# Візуалізація
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs = axs.ravel()

for i, (n, res) in enumerate(results.items()):
    axs[i].hist(res, bins=range(min(res), max(res)+2, 1), align='left', rwidth=0.8, edgecolor='black', alpha=0.7)
    axs[i].set_title(f'{n} симуляцій\nСереднє = {means[n]:.2f}')
    axs[i].set_xlabel('Ціна')
    axs[i].set_ylabel('Частота')

plt.tight_layout()
plt.show()

print("Середні значення:")
for n, mean in means.items():
    print(f"{n} симуляцій: {mean:.2f}")
```

### Інтерпретація результатів:

1. З збільшенням кількості симуляцій, середнє значення має наближатися до теоретичного 1.5.
2. Гістограми повинні показувати розподіл, який центрується навколо 1.5 з більшою кількістю симуляцій.
3. Розподіл має бути симетричним, оскільки ймовірності збільшення та зменшення ціни рівні.
